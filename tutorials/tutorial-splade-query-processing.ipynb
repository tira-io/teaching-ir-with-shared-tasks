```python
import pyterrier as pt
from tqdm import tqdm
import pandas as pd

if not pt.started():
  pt.init()

from tira.rest_api_client import Client
tira = Client()
```

# Dummy Evaluation on TREC DL 19

---



```python
dataset = pt.get_dataset("irds:msmarco-passage/trec-dl-2019/judged")
topics = dataset.get_topics()
topics.head(3)
```

```python
splade_reranker = tira.pt.from_retriever_submission('ir-benchmarks/naverlabseurope/Splade (re-ranker-scores-only)', dataset)
bm25 = tira.pt.from_retriever_submission('ir-benchmarks/tira-ir-starter/BM25 Re-Rank (tira-ir-starter-pyterrier)', dataset)
```

```python
pt.Experiment(
    [splade_reranker, bm25],
    names=['BM25 >> Splade (from TIRA Cache)', 'BM25'],
    topics=topics,
    qrels=dataset.get_qrels(),
    eval_metrics=['recip_rank', 'ndcg_cut.10']
)
```

```python
# this shows how the term importance/expansion of the splade query processing could be re-used / inspected
splade_query_processing = tira.pt.transform_queries('ir-benchmarks/naverlabseurope/Splade (re-ranker)', dataset, file_selection="*/queries.jsonl.gz")
splade_query_processing(topics).head(3)
```

```python
splade_query_processing(topics).iloc[0].to_dict()
```

# Evaluation on some more datasets

```python
ir_datasets = ['clueweb09/en/trec-web-2009', 'clueweb09/en/trec-web-2010', 'clueweb09/en/trec-web-2011',
               'clueweb09/en/trec-web-2012', 'clueweb12/trec-web-2013', 'clueweb12/trec-web-2014',
               'disks45/nocr/trec-robust-2004', 'antique/test', 'argsme/2020-04-01/touche-2020-task-1',
               'argsme/2020-04-01/touche-2021-task-1', 'clueweb12/touche-2020-task-2', 'clueweb12/touche-2021-task-2',
               'cord19/fulltext/trec-covid', 'cranfield', 'disks45/nocr/trec7', 'disks45/nocr/trec8',
               'gov/trec-web-2002', 'gov/trec-web-2003', 'gov2/trec-tb-2004',
               'gov2/trec-tb-2005', 'gov2/trec-tb-2006', 'medline/2004/trec-genomics-2004',
               'medline/2004/trec-genomics-2005', 'medline/2017/trec-pm-2017', 'medline/2017/trec-pm-2018',
               'msmarco-passage/trec-dl-2019/judged', 'msmarco-passage/trec-dl-2020/judged',
               'nfcorpus/test', 'vaswani', 'wapo/v2/trec-core-2018'
               ]
df_all = []

for dataset in tqdm(ir_datasets):
  pt_dataset = pt.get_dataset(f"irds:{dataset}")
  splade_reranker = tira.pt.from_retriever_submission('ir-benchmarks/naverlabseurope/Splade (re-ranker-scores-only)', pt_dataset)
  df = pt.Experiment(
      [splade_reranker],
      names=['BM25 >> Splade (from TIRA Cache)'],
      topics=pt_dataset.get_topics(),
      qrels=pt_dataset.get_qrels(),
      eval_metrics=['recip_rank', 'ndcg_cut.10']
  )
  df['dataset'] = dataset

  df_all += [df]
df_all = pd.concat(df_all)
```

```python
df_all.sort_values('ndcg_cut.10', ascending=False)
```
